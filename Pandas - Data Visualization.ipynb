```python
#Mapping top 10 high priced location
import folium

sorted_df = housing.sort_values(by='median_house_value', ascending=False)

# Select the top 10 high price places
top_10 = sorted_df.head(10)

# Create a folium map centered around the mean latitude and longitude
center_lat = top_10['latitude'].mean()
center_long = top_10['longitude'].mean()
map = folium.Map(location=[center_lat, center_long], zoom_start=5)

for index, row in top_10.iterrows():
    
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"Price: ${row['median_house_value']:.2f}",
    ).add_to(map)

# Display the map
map
```
